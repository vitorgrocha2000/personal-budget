```python
import os
import pandas as pd

df = pd.read_csv('transaction-history.csv')
```

```python
df = df.dropna(axis=1, how='all')
df = df[df['Situação'] != 'Cancelado']
df
```

```python

df_receitas = df[df['Direção'] == 'IN']
df_despesas = df[df['Direção'] == 'OUT']

# 3. Transformar os Dados
# Para receitas
df_receitas_transformado = pd.DataFrame({
    'Valor': df_receitas['Valor de destino (tarifas inclusas)'],
    'Efetuado': 1,  # Supondo que todas as receitas estão efetuadas
    'Fixo': 0,  # Supondo que todas as receitas não são fixas
    'Data': df_receitas['Concluída em'],
    'Categoria': 'Outros',  # Supondo que a categoria é 'Outros'
    'Descrição': df_receitas['Nome de origem']
})

# Para despesas
df_despesas_transformado = pd.DataFrame({
    'Valor': df_despesas['Valor de origem (tarifas inclusas)'],
    'Efetuado': 1,  # Supondo que todas as despesas estão efetuadas
    'Fixo': 0,  # Supondo que todas as despesas não são fixas
    'Data': df_despesas['Concluída em'],
    'Categoria': 'Outros',  # Supondo que a categoria é 'Outros'
    'Descrição': df_despesas['Nome do beneficiário']
})

# 4. Salvar os Novos CSVs
df_receitas_transformado.to_csv('receitas_transformadas.csv', index=False)
df_despesas_transformado.to_csv('despesas_transformadas.csv', index=False)
```
